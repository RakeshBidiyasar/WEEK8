```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering

X, y = make_moons(n_samples=300, noise=0.1, random_state=42)

dbscan = DBSCAN(eps=0.3, min_samples=5)
db_labels = dbscan.fit_predict(X)

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

agglo = AgglomerativeClustering(n_clusters=2)
agglo_labels = agglo.fit_predict(X)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

axs[0].scatter(X[:, 0], X[:, 1], c=db_labels, cmap='plasma', s=40)
axs[0].set_title("DBSCAN Clustering")
axs[0].set_xlabel("Feature 1")
axs[0].set_ylabel("Feature 2")

noise_mask = db_labels == -1
axs[0].scatter(X[noise_mask, 0], X[noise_mask, 1], c='black', marker='x', label='Noise')
axs[0].legend()

axs[1].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='plasma', s=40)
axs[1].set_title("K-Means Clustering")
axs[1].set_xlabel("Feature 1")

axs[2].scatter(X[:, 0], X[:, 1], c=agglo_labels, cmap='plasma', s=40)
axs[2].set_title("Hierarchical Clustering (Agglomerative)")
axs[2].set_xlabel("Feature 1")

plt.tight_layout()
plt.show()
```
